```python
#import nltk
#from nltk.tokenize import word_tokenize
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import string
import pandas as pd
file_path = 'C:\\Users\\namee\\OneDrive\\Desktop\\archive (2)\\Airline_review.csv'
airline_reviews_df = pd.read_csv(file_path)
airline_reviews_df.head()
```

```python
#lowercase
airline_reviews_df['Review_Title'] = airline_reviews_df['Review_Title'].str.lower()
airline_reviews_df['Review'] = airline_reviews_df['Review'].str.lower()
airline_reviews_df.head()
```

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
#tokenization
#'Review_Title' column
airline_reviews_df['Review_Title_Tokens'] = airline_reviews_df['Review_Title'].apply(word_tokenize)

#'Review' column
airline_reviews_df['Review_Tokens'] = airline_reviews_df['Review'].apply(word_tokenize)
airline_reviews_df.head()
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer


nltk.download('wordnet')

# Lemmatization
lemmatizer = WordNetLemmatizer()

#'Review_Title_Tokens' column
airline_reviews_df['Review_Title_Lemmas'] = airline_reviews_df['Review_Title_Tokens'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])

#'Review_Tokens' column
airline_reviews_df['Review_Lemmas'] = airline_reviews_df['Review_Tokens'].apply(lambda tokens: [lemmatizer.lemmatize(token) for token in tokens])
airline_reviews_df.head()
```
